#### Insight :

- Berdasarkan analisis 14 skala survei terdapat bottom 3 yang membuat penumpang **netral atau tidak puas** berdasarkan mean(rata-rata) dan 50%(median) yaitu `Inflight wifi service`, `Ease of Online booking` dan `Gate location`. Ketiga skala tersebut memiliki mean dibawah 3 dan memiliki median 3. 
- Dapat dilihat melalui `countplot` di atas, untuk top 3 *satisfied* banyak yang memberikan skor 4 dan 5. Namun, sebaliknya untuk bottom 3 *neutral or dissatisfied* lebih banyak pemberi skor 3, 2 serta 1.

#### Recommendation :
- Jaga top 3 skala yang membuat penumpang *satisfied* dan jika memungkinkan ditingkatkan lagi pelayanannya. Selain top 3 tersebut, sebenarnya skala dengan median = 4 dan mean lebih 3,3 sudah cukup memuaskan penumpang pesawat, sehingga harus dijaga juga pelayanannya.
- Untuk top 3 skala yang membuat penumpang *neutral or dissatisfied* harus ditingkatkan lagi pelayanannya. Bukan hanya top 3 tersebut, namun skala dengan median = 3 dan mean kurang 3,3 juga harus diperhatikan lagi agar tidak menjadi top 3 *neutral or dissatisfied*.
- Menurut sumber di bawah kenyamanan penumpang di pesawat dapat membuat penumpang merasa puas, seperti kenyamanan tempat duduk, hiburan di dalam pesawat.

#### **Business Question 5** 
Bagaimana hubungan umur dengan tingkat kepuasan penumpang?

```python
# Membuat range umur berdasarkan https://www.statcan.gc.ca/en/concepts/definitions/age2
def umur(x):
    if x <= 14 :
        return 'Children'
    elif x > 14 and x <=24 :
        return 'Youth'
    elif x > 24 and x <=64 :
        return 'Adult'
    else :
        return 'Senior'

df.insert(5, column='age_range',value=df['Age'].apply(umur))
```

```python
df['age_range'].value_counts()
```

```python
pd.crosstab(df['age_range'], df['satisfaction'])
```

```python
pd.crosstab(df['age_range'], df['satisfaction'], normalize='index').round(2)*100
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='age_range',hue='satisfaction', palette='PRGn')

#plt.savefig('b5.jpg')             
plt.show()
```

#### Insight :
Berdasarkan analisis kategori *Adult* paling banyak menggunakan jasa penerbangan dibandingkan kategori umur lain. Tingkat ketidakpuasan paling besar ada pada kategori *Children* sebesar 86% disusul oleh kategori usia *Senior* sebesar 82%. Sementara tingkat kepuasan paling besar ada pada kategori usia *Adult* sebesar 49% dan kategori *Youth* sebesar 32%.

#### Recommendation :
Untuk meningkatkan ketidakpuasan kategori *Adult* dan *Children* ada baiknya menyiapkan pelayanan khusus seperti selimut tambahan, makanan ringan, buku bacaan atau meningkatkan pelayanan pada skala survei seperti `Inflight entertainment`, `Seat comfort` serta `leg room service`
